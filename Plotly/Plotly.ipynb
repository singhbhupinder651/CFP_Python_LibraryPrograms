### 1. Write a program to draw a scatter plot for random 1000 x and y coordinates

```python
import pandas as pd
import numpy as np
from plotly.offline import iplot
import plotly as py
import plotly.tools as tls
import cufflinks as cf
```

```python
# if we want to execute plotly in offline mode
py.offline.init_notebook_mode(connected=True) 
cf.go_offline()
```

```python
df = pd.DataFrame(np.random.rand(1000,2), columns = ['a', 'b'])
df.head()
```

```python
cf.set_config_file(theme = 'polar')
df.iplot(x = 'a', y = 'b', mode = 'markers', kind = 'scatter',title = 'Scatter Plot',
        xTitle = 'X Axis', yTitle = 'Y Axis')
```

### 2. Write a program to draw line and scatter plots for random 100 x and y coordinates

```python
df = pd.DataFrame(np.random.rand(100,2), columns = ['a', 'b'])
df.head()
```

```python
df.iplot(x = 'a', y = 'b',mode = 'markers', kind = 'scatter',title = 'Scatter Plot',
         xTitle = 'X Axis', yTitle = 'Y Axis', bestfit=True)
```

### 3.Write a program to draw a scatter plot for random 500 x and y coordinates and style it

```python
df = pd.DataFrame(np.random.rand(500,2), columns = ['a', 'b'])
df.head()
```

```python
df.iplot(x = 'a', y = 'b', kind = 'scatter',title = 'Scatter Plot', xTitle = 'X Axis', 
         yTitle = 'Y Axis', mode = 'markers', color='LightBlue',size=15)
```

### 4.Write a program to draw a scatter plot for a given dataset and show datalabels on hover

https://raw.githubusercontent.com/plotly/datasets/master/2014_usa_states.csv

```python
import plotly.express as px
```

```python
df1 = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2014_usa_states.csv')
df1.head()
```

```python
px.scatter(df1, x = 'Rank', y = 'Population',title = 'Scatter Plot', hover_data= df1, color = 'State')
```
